```python
# Gerekli kütüphaneleri import edelim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Iris veri setini yükleyelim
data = load_iris()
X = data.data
y = data.target
```

```python
# Veriyi ölçeklendirelim (K-means algoritması mesafeye dayalı olduğu için önemlidir)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# K-means modelini k=3 (3 kümeye) olarak oluşturalım (Iris veri setinde 3 farklı tür var)
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python

# Tahmin edilen küme etiketleri
y_kmeans = kmeans.predict(X_scaled)

# Sonuçları görselleştirelim
plt.figure(figsize=(8, 6))
```

```python
# İlk iki özelliği kullanarak 2D görselleştirme yapalım
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Küme merkezlerini de gösterelim
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.75, marker='X')

plt.xlabel('Özellik 1')
plt.ylabel('Özellik 2')
```
